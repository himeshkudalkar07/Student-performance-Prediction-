```python
# Student Performance Prediction using Python & Machine Learning
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
# Load dataset
df = pd.read_csv("student_performance.csv")

print("Dataset Preview:")
print(df)

print("\nDataset Info:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())
# Convert target column to numeric
df['final_result'] = df['final_result'].map({'Pass':1, 'Fail':0})
# Separate features and target
X = df[['study_hours','attendance','previous_score']]
y = df['final_result']
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# Train Logistic Regression model
model = LogisticRegression()
model.fit(X_train, y_train)
# Make predictions
y_pred = model.predict(X_test)
# Model accuracy
accuracy = accuracy_score(y_test, y_pred)
print("\nModel Accuracy:", accuracy)
# Try custom prediction
sample_prediction = model.predict([[6,80,65]])
print("\nPrediction for [6 study hours, 80 attendance, 65 previous score]:",
      "Pass" if sample_prediction[0] == 1 else "Fail")
# Visualization: Study Hours vs Result
df.groupby('final_result')['study_hours'].mean().plot(kind='bar')
plt.title("Average Study Hours by Result")
plt.xlabel("Result (0=Fail, 1=Pass)")
plt.ylabel("Average Study Hours")
plt.show()
# Visualization: Attendance vs Result
df.groupby('final_result')['attendance'].mean().plot(kind='bar')
plt.title("Average Attendance by Result")
plt.xlabel("Result (0=Fail, 1=Pass)")
plt.ylabel("Average Attendance")
plt.show()

print("\nKey Learning:")
print("- Higher study hours and attendance increase pass probability")
print("- Logistic Regression is suitable for binary classification problems")
print("- This model helps predict student performance using basic features")
```
